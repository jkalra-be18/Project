```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Administrator\Desktop\Finance_europe.csv')
df.head()
df.set_index('Country')
df.drop(df.index[27])
```

```python
df.set_index('Country')
```

```python
df = df.drop(df.index[27])
```

```python
fig = plt.figure(figsize = (10,6))
sns.barplot(y = df['Country'], x = df['With difficulty'], color = 'b', label = 'With difficulty')
sns.barplot(y = df['Country'], x = df['With great difficulty'], color ='r', label = 'With great difficulty')
plt.xlabel('Difficulty Reviews', size = '15')
plt.ylabel('Countries of Europe', size = '15')
plt.title('Is your household able to make ends meet? by country', size = '15')
plt.legend()
plt.show()
```
